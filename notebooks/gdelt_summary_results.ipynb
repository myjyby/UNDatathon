```python
import pandas as pd
from upath import UPath as Path
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
files_dir = Path("/Users/ewheeler/dev/UNDatathon/data/gdelt_files/")
```

```python
target_countries_iso = ['UKR', 'DJI', 'EGY', 'KEN', 'MOZ', 'SDN', 'TZA', 'RUS']
```

```python
df = pd.read_json(files_dir / "models_summary.json").T
df
```

```python
_df = pd.read_feather(files_dir / 'gdelt_monthly_fpi.feather')
_df = _df[_df['iso3'].isin(target_countries_iso)]
_df['fpi_pred'] = ''
_df['fpi_pred_orig'] = ''
_df
```

```python
train_months = _df.query("Date < '2022-07-01'")['Date'].unique().strftime('%Y-%m-%d')
print(train_months[0], len(train_months))
val_months = _df.query("'2022-07-01' <= Date < '2023-01-01'")['Date'].unique().strftime('%Y-%m-%d')
print(val_months[0], len(val_months))
test_months = _df.query("Date >= '2023-01-01'")['Date'].unique().strftime('%Y-%m-%d')
print(test_months[0], len(test_months))
```

```python
test_losses = list()
for mod in df.iterrows():
    country = mod[0]
    test_mse = mod[1][3]
    test_mse_orig_units = mod[1][4]
    test_losses.append((country, test_mse, test_mse_orig_units))
    y_test_pred = mod[1][11]
    y_test_pred_orig_units = mod[1][12]
    for i, m in enumerate(test_months[:-2], 0):
        print(i, m)
        _df.at[f"{m}_{country}", "fpi_pred"] = y_test_pred[i]
        _df.at[f"{m}_{country}", "fpi_pred_orig"] = y_test_pred_orig_units[i]
```

```python
_df.to_csv(files_dir / "gdelt_models_predictions.csv")
```
